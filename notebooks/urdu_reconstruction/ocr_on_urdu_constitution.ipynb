```python
@dataclass
class Options:
    saved_model: str
    character: str
    device: str
    batch_max_length: int = 100
    imgH: int = 32
    imgW: int = 400
    rgb: bool = False
    FeatureExtraction: str = "HRNet"
    SequenceModeling: str = "DBiLSTM"
    Prediction: str = "CTC"
    num_fiducial: int = 20
    input_channel: int = 1
    output_channel: int = 32
    hidden_size: int = 256
    device_id: str = None
    num_class = None
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print("Device:", device)
```

```python
with open("UrduGlyphs.txt", "r", encoding="utf-8") as f:
    content = f.readlines()

content = ''.join([str(elem).strip('\n') for elem in content])

# Instantiate the options
opt = Options(
    saved_model="best_norm_ED.pth",
    character=content + " ",
    device=device
)
```

```python
# Model configuration
if 'CTC' in opt.Prediction:
    converter = CTCLabelConverter(opt.character)
else:
    converter = AttnLabelConverter(opt.character)
opt.num_class = len(converter.character)

if opt.rgb:
    opt.input_channel = 3

# Load model
model = Model(opt)
model = model.to(device)

model.load_state_dict(torch.load(opt.saved_model, map_location=device))
model.eval()

detection_model = YOLO("yolov8m_UrduDoc.pt")
```

```python
def detect(img):
    if isinstance(img, str):
        input = Image.open(img)
    else:
        input = img

    detection_results = detection_model.predict(source=input, conf=0.2, imgsz=1280, save=False, nms=True, device=device)
    bounding_boxes = detection_results[0].boxes.xyxy.cpu().numpy().tolist()
    # bounding_boxes.sort(key=lambda x: x[1])

    return bounding_boxes
```

```python
def crop_image(img, bounding_boxes):
    x1, y1, x2, y2 = bounding_boxes
    cropped_img = img.crop((x1, y1, x2, y2))

    return cropped_img
```

```python
def predict_text(img):
    # Load image
    if isinstance(img, str):
        img = Image.open(img)

    if opt.rgb:
        img = img.convert('RGB')
    else:
        img = img.convert('L')

    img = img.transpose(Image.Transpose.FLIP_LEFT_RIGHT)

    # Resize image
    w, h = img.size
    ratio = w / float(h)
    if math.ceil(opt.imgH * ratio) > opt.imgW:
        resized_w = opt.imgW
    else:
        resized_w = math.ceil(opt.imgH * ratio)
    img = img.resize((resized_w, opt.imgH), Image.Resampling.BICUBIC)

    # Normalize image
    transform = NormalizePAD((1, opt.imgH, opt.imgW))
    img = transform(img)
    img = img.unsqueeze(0)  # Add batch dimension

    batch_size = img.shape[0]
    img = img.to(device)

    # Prediction
    preds = model(img)
    preds_size = torch.IntTensor([preds.size(1)] * batch_size)

    _, preds_index = preds.max(2)
    preds_str = converter.decode(preds_index.data, preds_size.data)[0]

    return preds_str
```

```python
class UnionFind:
    def __init__(self, n):
        self.parent = list(range(n))
        self.rank = [1] * n

    def find(self, x):
        if self.parent[x] != x:
            self.parent[x] = self.find(self.parent[x])
        return self.parent[x]

    def union(self, x, y):
        rootX = self.find(x)
        rootY = self.find(y)
        if rootX != rootY:
            if self.rank[rootX] > self.rank[rootY]:
                self.parent[rootY] = rootX
            elif self.rank[rootX] < self.rank[rootY]:
                self.parent[rootX] = rootY
            else:
                self.parent[rootY] = rootX
                self.rank[rootX] += 1

def merge_bboxes_optimized(detections, threshold):
    def is_same_line(bbox1, bbox2, threshold):
        # Check if two bboxes are on the same line based on vertical proximity
        _, y1_min, _, y1_max = bbox1
        _, y2_min, _, y2_max = bbox2
        return abs(y1_min - y2_min) < threshold or abs(y1_max - y2_max) < threshold

    def merge_bbox(bbox1, bbox2):
        # Merge two bboxes by finding the min/max coordinates
        x_min = min(bbox1[0], bbox2[0])
        y_min = min(bbox1[1], bbox2[1])
        x_max = max(bbox1[2], bbox2[2])
        y_max = max(bbox1[3], bbox2[3])
        return (x_min, y_min, x_max, y_max)

    # Step 1: Sort the detections by the y_min of the bbox
    detections = sorted(detections, key=lambda d: d["bbox"][1])

    n = len(detections)
    uf = UnionFind(n)

    # Step 2: Compare each bbox with others in its vicinity (sweeping)
    for i in range(n):
        for j in range(i + 1, n):
            bbox1 = detections[i]["bbox"]
            bbox2 = detections[j]["bbox"]

            # Since the list is sorted, if bbox2's y_min is outside the threshold of bbox1, stop comparison
            if bbox2[1] - bbox1[3] > threshold:
                break

            # If they are vertically within the threshold, check if they are on the same line
            if is_same_line(bbox1, bbox2, threshold):
                uf.union(i, j)

    # Step 3: Merge all bounding boxes in the same group
    merged_groups = {}
    for i in range(n):
        root = uf.find(i)
        if root not in merged_groups:
            merged_groups[root] = detections[i]["bbox"]
        else:
            merged_groups[root] = merge_bbox(merged_groups[root], detections[i]["bbox"])

    # Step 4: Return the merged bounding boxes
    return [{"bbox": bbox} for bbox in merged_groups.values()]
```

```python
def get_page_wise_detections(ocr_detections, thresh=5):
    page_detections = {}

    for detection in ocr_detections:
        page_no = detection["page_no"]

        try:
            page_detections[page_no].append(detection)
        except:
            page_detections[page_no] = [detection]

    for page_no, detections in tqdm(page_detections.items()):
        page_detections[page_no] = merge_bboxes_optimized(detections, thresh)

    return page_detections
```

```python
from reportlab.pdfgen import canvas
from reportlab.pdfbase.ttfonts import TTFont
from reportlab.pdfbase import pdfmetrics
from reportlab.lib.pagesizes import letter
import arabic_reshaper
from bidi.algorithm import get_display
from pdf2image import convert_from_path
import json
from tqdm.notebook import tqdm
import numpy as np
from shapely.geometry import box
```

```python
def bbox_iou(box1, box2):
    b1 = box(*box1)
    b2 = box(*box2)

    inter_area = b1.intersection(b2).area
    union_area = b1.union(b2).area

    iou = inter_area / union_area if union_area != 0 else 0
    return iou

def process_page_detections(detections, thresh):
    unique_detections = []
    seen_texts = {}

    for detection in detections:
        text = detection["text"]
        bbox = detection["bbox"]

        if text not in seen_texts:
            seen_texts[text] = []

        to_add = True
        for existing_bbox in seen_texts[text]:
            if bbox_iou(bbox, existing_bbox) > thresh:
                to_add = False
                break

        if to_add:
            seen_texts[text].append(bbox)
            unique_detections.append(detection)

    return unique_detections

def process_detections(page_detections, thresh=0.7):
    processed_detections = {}

    for page_no, detections in tqdm(page_detections.items()):
        processed_detections[page_no] = process_page_detections(detections, thresh)

    return processed_detections
```

```python
processed_detections = process_detections(ocr_detections, thresh=0.2)
```

```python
def draw_arabic_text(c, text, bbox, page_size):
    reshaped_text = arabic_reshaper.reshape(text)
    display_text = get_display(reshaped_text)
    x0, y0, x1, y1 = bbox
    width, height = page_size

    # Convert coordinates from image space to PDF space
    x0 = x0 / width * letter[0]
    y0 = (height - y1) / height * letter[1]

    c.setFont('NotoNaskh', 12)
    c.drawString(x0, y0, display_text)
```

```python
# Register a font that supports Arabic
pdfmetrics.registerFont(
    TTFont('NotoNaskh', '/content/NotoNaskhArabic-Regular.ttf'))

# Create the PDF
c = canvas.Canvas(OUTPUT_PATH, pagesize=letter)

# Set the font to the one that supports Arabic
c.setFont('NotoNaskh', 20)
```

```python
PADDING = 20
padded_page_size = (letter[0] + PADDING, letter[1] + PADDING)
```

```python
page_sizes = [img.size for img in pdf_images]
```

```python
for page_no in tqdm(range(len(page_sizes))):
    c.setPageSize(padded_page_size)

    try:
        dets = processed_detections[page_no]

        for det in dets:
            draw_arabic_text(c, det['text'], det['bbox'], page_sizes[page_no])
    except:
        pass
    finally:
        c.showPage()
```
