````python
import PyPDF2

def find_backtick_pairs(pdf_path):
    backtick_pairs = []
    backtick_count = 0
    odd_occurrence_page = None
    
    with open(pdf_path, 'rb') as pdf_file:
        reader = PyPDF2.PdfReader(pdf_file)
        num_pages = len(reader.pages)
        
        for page_num in range(num_pages):
            page_text = reader.pages[page_num].extract_text()
            
            # Count occurrences of triple backticks in this page
            occurrences = page_text.count('```')
            
            for _ in range(occurrences):
                backtick_count += 1
                
                if backtick_count % 2 == 1:
                    odd_occurrence_page = page_num + 1
                else:
                    # If it's an even occurrence, pair it with the previous odd page number
                    if odd_occurrence_page is not None:
                        backtick_pairs.append((odd_occurrence_page, page_num + 1))
                    odd_occurrence_page = None 
    return backtick_pairs

# Example usage
pdf_path = '../../data/constitution_pdfs/constitution_english_reduced_delimited.pdf'
result = find_backtick_pairs(pdf_path)
````
