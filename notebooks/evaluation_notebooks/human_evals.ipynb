```python
import pandas as pd
```

```python
models = [
    "claude-3.5-sonnet",
    "gemini-flash",
    "gemini-pro",
    "mistral-large",
    "mt5-large"
]
```

```python
dfs = []

for i in models:
    df = pd.read_excel(f"../../data/human_evaluations_qa/{i}_human_eval.xlsx")
    dfs.append(df)
```

```python
for i in range(len(models)):
    curr_df = dfs[i]
    if "score" in curr_df.columns:
        curr_df.rename(columns={"score": "Score"}, inplace=True)
```

```python
for i in dfs:
    display(i.head(3))
```

```python
for i in range(len(models)):
    curr_df = dfs[i]
    print("Model:", models[i])
    print(100 * sum(curr_df["Score"]) / len(curr_df["Score"]))
```
