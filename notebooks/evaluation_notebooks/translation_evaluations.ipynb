```python
import evaluate
import pandas as pd
```

```python
metric = evaluate.load("sacrebleu")
metric_2 = evaluate.load("chrf")
```

```python
models = [
    "claude",
    "gemini-1.5-flash-002",
    "gemini-1.5-pro-002",
    "mistral",
    "IndicTrans2"
]
```

```python
model_responses = [pd.read_csv(f"../../data/model_translation_responses/context_translations_{model}.csv") for model in models]
```

```python
for i in model_responses:
    display(i.head())
```

```python
references = list(model_responses[0].iloc[:, 1].values)
```

```python
references[:3]
```

```python
for i in range(len(model_responses)):
    predictions = model_responses[i].iloc[:, 2]
    print(f"Model: {models[i]}")
    bleu_result = metric.compute(predictions=predictions, references=references)
    chrf_result = metric_2.compute(predictions=predictions, references=references, word_order=2)
    print(f"SacreBLEU: {bleu_result['score']}")
    print(f"CHRF: {chrf_result['score']}")
    print("--------------------")
```
