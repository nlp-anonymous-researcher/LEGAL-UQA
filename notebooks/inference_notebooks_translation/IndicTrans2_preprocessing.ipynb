```python
import pickle
```

```python
# Load the list of tuples from the pickle file
with open('../../data/model_translation_responses/IndicTrans2_translations.pkl', 'rb') as file:
    translations = pickle.load(file)
```

```python
# Calculate word count and store sentences with their original indices
sentences_with_length_and_index = [
    (i, sentence[0], len(sentence[0].split(' '))) for i, sentence in enumerate(translations)
]

# Sort by word count
sorted_sentences = sorted(sentences_with_length_and_index, key=lambda x: x[2])

# Get the 50 smallest sentences
top_50_smallest_sentences = sorted_sentences[:50]
smallest_context_indices = [sentence[0] for sentence in top_50_smallest_sentences]
```

```python
smallest_translations = [translations[i] for i in smallest_context_indices]
smallest_translations
```

```python
from datasets import load_from_disk

dataset_dict = load_from_disk("../../data/complete_dataset/")
```

```python
y = dataset_dict['validation']['context_urdu']
x = dataset_dict['validation']['context_eng']
```

```python
import pandas as pd

# Create DataFrame from smallest_translations
df_translations = pd.DataFrame(smallest_translations, columns=['context_eng', 'predicted_urdu_IndicTrans2'])

# Create DataFrame from x and y lists
df_validation = pd.DataFrame({
    'context_eng': x,
    'context_urdu': y
})

# Merge the DataFrames on 'context_eng' and 'context_urdu' to find common rows
final_df = df_validation.merge(df_translations, on='context_eng', how='inner')

# Display the final DataFrame
display(final_df)
```

```python
final_df.to_csv('../../data/model_translation_responses/IndicTrans2_translations.csv', index=False)
```

```python
ground_truths = final_df[['context_eng', 'context_urdu']]
```
